# Puzzle - part 1

**--- Day 3: Lobby ---**

You descend a short staircase, enter the surprisingly vast lobby, and are quickly cleared by the security checkpoint.\
When you get to the main elevators, however, you discover that each one has a red light above it: they're all **offline**.

"Sorry about that," an Elf apologizes as she tinkers with a nearby control panel.\
"Some kind of electrical surge seems to have fried them. I'll try to get them online soon."

You explain your need to get further underground.\
"Well, you could at least take the escalator down to the printing department,\
not that you'd get much further than that without the elevators working.\
That is, you could if the escalator weren't also offline."

"But, don't worry! It's not fried; it just needs power.\
Maybe you can get it running while I keep working on the elevators."

There are batteries nearby that can supply emergency power to the escalator for just such an occasion.\
The batteries are each labeled with their **joltage** rating, a value from `1` to `9`.\
You make a note of their joltage ratings (your puzzle input). For example:
```
987654321111111
811111111111119
234234234234278
818181911112111
```

The batteries are arranged into **banks**; each line of digits in your input corresponds to a single bank of batteries.\
Within each bank, you need to turn on exactly **two batteries**;\
the joltage that the bank produces is equal to the number formed by the digits on the batteries you've turned on.\
For example, if you have a bank like `12345` and you turn on batteries `2` and `4`, the bank would produce 24 jolts.\
(You cannot rearrange batteries.)

You'll need to find the largest possible joltage each bank can produce. In the above example:

- In `987654321111111`, you can make the largest joltage possible, `98`, by turning on the first two batteries.
- In `811111111111119`, you can make the largest joltage possible by turning on the batteries labeled `8` and `9`, producing 89 jolts.
- In `234234234234278`, you can make `78` by turning on the last two batteries (marked `7` and `8`).
- In `818181911112111`, the largest joltage you can produce is `92`.

The total output joltage is the sum of the maximum joltage from each bank,\
so in this example, the total output joltage is `98` + `89` + `78` + `92` = `357`.

There are many batteries in front of you.\
Find the maximum joltage possible from each bank; **what is the total output joltage?**

## Input

````python
# Load the input file

with open('input - Day 3.txt', 'r') as file:
    file_input = file.read()


print(file_input[:504])
````

## Input Formatting

Let's extract each line first

````python
from pprint import pprint

input_lines = file_input.split('\n')
pprint(input_lines[:5])

print(f"\nThere are {len(input_lines)} lines/banks")
````

Make sure the last line is not empty

````python
print(f"Last line: {input_lines[-1]}")
# Check the last line, if it's empty then let's remove it
if input_lines[-1] == "":
    del input_lines[-1]

print(len(input_lines))
````

Now let us conver each line (bank) to a list of numbers so that we can iterate over them and compare them.


````python
banks = []

for line in input_lines:
    bank = [ int(char) for char in list(line) ]
    banks.append(bank)

print(banks[0])
````

## Solution

This seems relatively simple we want to find the largest number first.\
Then the largest number after the one we just found.

One important caveat, the first battery cannot be the last number,\
because then we will have no options left for the second battery

Let's not overcomplicate this and simply use two loops.

````python
all_voltage = []

for bank in banks:
    # First loop - we could use numpy but our input
    # is small enough so we will just type out a loop
    first_battery = (-1, None) # (voltage, index)
    for idx, battery_voltage in enumerate(bank):
        # Last battery cannot be the first battery
        if idx == len(bank) - 1:
            continue

        # Note we prefer first batter found so we use > not >=
        # That is so that we will have more options left for the second_battery
        if battery_voltage > first_battery[0]:
            first_battery = (battery_voltage, idx)

    # Second loop - we start from the index of the first_battery
    # and simply repeat the entire process
    second_battery_voltage = -1
    start_idx = first_battery[1] + 1
    for battery_voltage in bank[start_idx:]:

        if battery_voltage > second_battery_voltage:
            second_battery_voltage = battery_voltage

    voltage = int(str(first_battery[0]) + str(second_battery_voltage))
    all_voltage.append(voltage)
````

````python
print(f"Total joltage is {sum(all_voltage)} jolts!")
````

## Solution

Very similar to what we did already.\
Except now, we need to restrict last `n` batteries for each battery.

Let's simplify, say we only want to choose `5` batteries
If we are looking for the first battery,\
then the last possible battery we choose cannot leave us less than `4` batteries after it.\
For example in `4134189142` even though `9` is the highest number we'd only get `9142`.

Once again if we have multiple options for a battery, we choose the first one (from left).\
So in `3213900098888`, we will choose `9` then `9` then `8`, `8`, `8`.

Also notice that since we are assembling a decimal from left to right,\
we always want to prioritize the best batteries in that order.

In summary same as before but we need to dynamically restrict our searchable range.
Figuring out indices might be a bit tricky.

````python
all_voltages = []

for bank in banks:
    batteries_volt = ""
    current_battery_idx = 0

    for battery_number in range(12): # [0, 1, 2 ... 11]

        # Search range
        lower_bound = current_battery_idx
        if battery_number > 0:
            lower_bound += 1 # start from the next battery
        upper_bound = (len(bank) - 1) - 11 + battery_number #len != idx
        search_indices = [x + lower_bound for x in list(range(upper_bound + 1 - lower_bound))] # range is exclusive of stop value
        # print(search_indices) # For debugging

        # Check each battery in the bank
        best_volt = -1
        for idx in search_indices:
            battery_volt = bank[idx]

            if best_volt < battery_volt:
                best_volt = battery_volt
                current_battery_idx = idx

        batteries_volt += str(best_volt)

    all_voltages.append(int(batteries_volt))
````

````python
print(f"Total joltage is {sum(all_voltages)} jolts!")
````
